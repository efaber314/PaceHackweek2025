```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Bounding box coordinates (WSEN)
west, south, east, north =  13.3830, 29.5381, -33.4435, -10.4448

# Set up the map
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.stock_img()
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':', linewidth=0.5)

# Plot bounding box as a rectangle
ax.plot(
    [west, east, east, west, west],
    [south, south, north, north, south],
    color='red', linewidth=2, transform=ccrs.PlateCarree(), label='Bounding Box'
)

# Set extent slightly larger than bounding box for better visibility
ax.set_extent([west - 5, east + 5, south - 5, north + 5])

plt.title("Bounding Box Location")
plt.legend()
plt.show()
```
